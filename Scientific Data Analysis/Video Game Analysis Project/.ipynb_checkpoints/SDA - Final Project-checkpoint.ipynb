# 1. Introduction

### A. The Project
The goal of this project is to test whether a rating from a game review website has a direct relationship to it's sales. We will be using American game reviews to compare with global sales data broken down into regions. Some of the big questions we're trying to answer are: Does an American game review affect sales in North America/Europe/Japan?, Does a good review always mean good sales?, Does a bad review always mean bad sales?, Do reviewers favour a specific Platform?, and What games are the greatest of all time (best review and sales). 


### B. The Datasets
The first datasets we've chosen to use for this analysis is '20 years of games', a csv file which contains IGN review data from 2012 to 1998. The main highlight of IGN reviews is their scoring system which works on a scale from 0 to 10, with 0 being the lowest. IGN is video game media hub website for various videos and articles on all types of games. Their videos can range from reviews, gameplay, guides, or even promotional videos which act as ads. The other dataset we've chose to use is 'Video game sales', a csv file containing data on video game sales by region from 1980 to 2017. Video games use a similar system to DVD's region codes, such that each region is locked out from using other region's content. The 4 sales regions are: North America, Europe, Japan, and Other.

# 2. Preprocessing

### A. Importing Data & Packages
The code below imports a few packages we need to use for our analyzation of our datasets. There is also code for reading data from the '20 years of games' and 'Video game sales' CSV files. The data is stored in a Pandas dataframe.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as pyplot
%matplotlib inline

reviews = pd.read_csv('ign.csv', names=['UID', 
                                  'score_phrase', 
                                  'title', 
                                  'url', 
                                  'platform', 
                                  'score',
                                  'genre',
                                  'editors_choice',
                                  'release_year',
                                  'release_month',
                                  'release_day'])

sales = pd.read_csv('vgsales.csv', names=['Rank', 
                                        'Name', 
                                        'Platform', 
                                        'Year', 
                                        'Genre',
                                        'Publisher',
                                        'NA_Sales',
                                        'EU_Sales',
                                        'JP_Sales',
                                        'Other_Sales',
                                        'Global_Sales'])
```

# 3. Processing

## Total Game Reviews per Year

### A. Code Process Walkthrough : Total Game Reviews per Year
In the code below we wanted to know how many game reviews come out for each specific
year. We did this by first creating a list of reviews and organized them by year. We found out that
there were outliers where there was only 1 review for the year. We decided to omit that to make
the graph more readable and provide a better overall picture.


### B. Total Game Reviews per Year Code & Graph

```python
#total game reviews
#create list of reviews per year
reviews_per_year=reviews.copy().groupby('release_year')['UID'].nunique().tolist()

#create a list of unique years
unique_years=reviews.copy().release_year.unique().tolist()

#remove 'release_year' title and data from list and sort it
del unique_years[0]
del reviews_per_year[len(reviews_per_year)-1]

#sort and remove '1970' title and data
unique_years=sorted(list(map(int, unique_years)))
del unique_years[0]
del reviews_per_year[0]

#create pandas dataframe
reviews_per_year = pd.DataFrame({'Year': unique_years,'Reviews_Published':reviews_per_year},columns=['Year','Reviews_Published']).set_index('Year')

#plot graph
reviews_per_year.plot.bar(figsize=(len(unique_years), 5),title="Total Game Reviews per Year")
```

### C. Insights – Game Reviews Per Year
Based on the data we’ve received the year where video games have the most reviews is 2008. Likewise, the lowest amount of video game reviews for the year is 1996. The reason why 1996 is the lowest amount of reviews has to do with the review company from which the data originates from, IGN. At that point in time, IGN was just founded and had little employees to review all the games that were coming out. With that knowledge it makes sense that 1996 has the lowest amount of reviews. After the founding the increase of video game reviews have increased slowly increased and peaked in 2008. However,  the amount of reviews after that year have slowly decreased. We suspect that this decrease is due to the subprime mortgage crisis where thousands of people and businesses, which include the video game industry and IGN, lost a lot of money and jobs in the market. Due to shortage of money and cash-flow only a handful of video game businesses survived aftermath and only a handful of games were
able to successfully launch. That would explain why there are only a few video-game publishers
today.


### D. Summary – Game Reviews Per Year
* 1996 – Lowest amount reviews for the year
* 2008 – Highest amount of reviews for the year
* Steadily increase of video games reviews up until 2008
* Years after 2008 the amount of reviews have decreased substantially

### E. Code Process Walkthrough: Total Game Genre Releases
In the code below we wanted to know the breakdown of video genre’s that are being released and more importantly how many of each genre have been released so far. So what we did was first we created a list of genres that each video game was categorized in and made sure that it was unique to prevent duplicates. We then kept this count and used a bar-graph to visualize the breakdown of  video game genres and how many have been released in its category.

### F. Total Game Genre Releases Code & Graph

## Total Sales per Year by Region

### I. Code Process Walkthrough: Total Sales per Year by Region
In the code below we first created 3 different types of sales lists. Each list was broken down into the 3 regions: North American, Japanese and European. Afterwards, we removed bad data and “useless” data-entries in 2017 since the amount of data-points were limited when compared to the years before it. Afterwards, we plotted the data using a stacked bar-chart to make it easy to compare sales per year by region.


### J. Total Sales per Year by Region Code & Graph

```python
#total sales (in millions) per year broken down by region
#create list of sales per year
global_sales_per_year=sales.copy().groupby('Year')['Global_Sales'].unique()
NA_sales_per_year=sales.copy().groupby('Year')['NA_Sales'].unique()
JP_sales_per_year=sales.copy().groupby('Year')['JP_Sales'].unique()
EU_sales_per_year=sales.copy().groupby('Year')['EU_Sales'].unique()

#remove bad/useless data entries
del global_sales_per_year['Year']
del global_sales_per_year['2020']
del global_sales_per_year['2017']
del NA_sales_per_year['Year']
del NA_sales_per_year['2020']
del NA_sales_per_year['2017']
del JP_sales_per_year['Year']
del JP_sales_per_year['2020']
del JP_sales_per_year['2017']
del EU_sales_per_year['Year']
del EU_sales_per_year['2020']
del EU_sales_per_year['2017']

#create a list of unique years
unique_years=[x for x in sales.copy().Year.unique().tolist() if str(x) != 'nan']
unique_years=sorted(unique_years)

#remove bad/useless data entries
del unique_years[len(unique_years)-1]
del unique_years[len(unique_years)-1]
del unique_years[len(unique_years)-1]

#plot the chart
pyplot.figure(figsize=(20, 5))
pyplot.ylabel("Sales in Millions")
pyplot.xlabel("Year")
pyplot.title("Total Sales per Year per Region")

#for each year in global sales sum each region and set it as the bottom of the column, creating a stacke chart
for idx, val in enumerate(global_sales_per_year):
    bottom_data=0
    NA_sum=sum(map(float, NA_sales_per_year[idx]))
    p1=pyplot.bar(unique_years[idx], NA_sum, align='center', width=0.5, bottom=bottom_data,color='r')
    bottom_data = NA_sum
    
    JP_sum=sum(map(float, JP_sales_per_year[idx]))
    p2=pyplot.bar(unique_years[idx],JP_sum, align='center', width=0.5, bottom=bottom_data,color='g')
    bottom_data += JP_sum
    
    p3=pyplot.bar(unique_years[idx],sum(map(float, EU_sales_per_year[idx])), align='center', width=0.5, bottom=bottom_data,color='b')
    pyplot.legend((p1[0], p2[0], p3[0]), ('NA', 'JP', 'EU'))
```

### K. Insights – Total Sales (in millions) per year broken down by region
Looking at this data we can see the most video game sales take place in the North American region followed by Europe. Additionally, we can also see that video game sales have been steadily increasing year after year. The reason why video game sales have been steadily increasing was due to the release of new gaming consoles such as the NES, SNES, Gameboy and PlayStation 2 which made it easy for consumers to play video games. Due to the massive success of the hardware, Nintendo released several other video-game hardware devices in the subsequent years such as NES & SNES and eventually caught up to the North American Region in terms of sales in 1992 with the release of a video game titled “A Link to the Past”. Due to the lack of data-point in 2016, no real analysis can be applied and is considered an outlier.


### L. Summary - Total Sales (in millions) per year broken down by region
*	Overall most of the video game sales take place in the North American region followed by Europe

## Top 10 Publlishers by Number of Unique Releases

### M. Code Process Walkthrough: Top 10 Publlishers by Number of Unique Releases
To find out what the top 10 publishers are in the video-game industry we first created a list of reviews per year. Afterwards we sorted them out by publishers and sorted them by descending order. Afterwards, we limited the results to show only first 10 results. Since the results are sorted the 10 results in the data-frame are the top publishers.


### N. Top 10 Publlishers by Number of Unique Releases Code & Graph

```python
#Top 10 Publishers by Release
#create list of reviews per year
list_of_publishers=sales.copy().Publisher.unique().tolist()

#delete Title row
del list_of_publishers[0]

#sort and convert list to strings (some publishers have numbers in their name)
list_of_publishers=sorted(map(str,list_of_publishers))

#create pandas dataframe
top_ten_publishers = pd.DataFrame({'Publisher_name': list_of_publishers,
                                'Games_published':sales.copy().groupby('Publisher')['Rank'].nunique().tolist()}, 
                                columns=['Publisher_name','Games_published'])

#sort by games published and reset the indexing
top_ten_publishers=top_ten_publishers.sort_values('Games_published',ascending=False).reset_index(drop=True)

#plots graph
top_ten_publishers.head(10).plot.bar(x='Publisher_name',y='Games_published',title="Top 10 Publishers by Release")
```

## Most Sales of a Certain Platform

### Q. Code Process Walkthrough: Most Sales of a Certain Platform
To determine which video game platform had the most sales we created a list of reviews per year created a list of platforms from that list. Afterwards, we used the global sales data that the data-set had and summed each global sales data for each video game with respect to the platform it was built in. We then plotted this data using a bargraph.

### R. Most Sales of a Certain Platform Code & Graph

```python
#most sales of a certain platform
#create list of reviews per year
list_of_platforms=sales.copy().Platform.unique().tolist()

#delete title row
del list_of_publishers[0]

#create list of platforms and their sales
sales_per_platform=sales.copy()[['Platform','Global_Sales']]

#delete title row
sales_per_platform.drop(0,inplace=True)
                       
#convert global sales to floats
sales_per_platform['Global_Sales']=sales_per_platform['Global_Sales'].astype(float)
                       
#sum all global sales by platform
sales_per_platform = sales_per_platform.groupby(['Platform']).sum()

#plot graph
sales_per_platform.sort_values('Global_Sales',ascending=False).plot.bar(figsize=(20, 5),title="Total Sales by Platform (in Millions)")
```

### S. Insights - Most sales of a video game platform
In the graph we can clearly see that the PlayStation 2 has the most sales followed by Xbox 360. The successor, the PlayStation 3, could not match global sales record due to the 2008 sub-prime mortgage crisis. What’s interesting is the largest bars in the graph come from platforms produced by only 3 companies: Sony, Microsoft and Nintendo. 


### T. Summary - Most sales of a video game platform 
* PlayStation 2 has the most sales of a video game platform followed by Xbox 360.
* PlayStation 3, the successor of the PlayStation 2, did not 
* Sony, Microsoft and Nintendo dominate the video-game platform sales


## Average Review Score by Platform

### U. Code Process Walkthrough: Average Review Score by Platform 
To find out the average reviews score by platform we first created a copy of the review data-set and filtered out only the scores from that list. Afterwards, we then aggregated the list of reviews by platform, title and score and plotted the data using a bar - graph in descending values.


### V. Average Review Score by Platform Code & Graph

```python
#average review scores by platform 
list_of_reviews = reviews.copy()

#drop title row
list_of_reviews.drop(0,inplace=True)

#convert score values to floats
list_of_reviews['score']=list_of_reviews['score'].astype(float)

#remove excess columns
list_of_reviews = list_of_reviews[['title','score','platform']]

#find average score of each platform, then sort by average score
list_of_reviews = list_of_reviews.groupby(by=['platform'])['score'].mean().sort_values(ascending=False)

#plot graph with descending values
list_of_reviews.plot(kind='bar',figsize=[20,5],title="Average Review Score by Platform")
```

### W. Insights - Average Review scores by platform and by publisher
Looking at the bar-graph above we can see that SteamOS has the highest average score. We believe this is the case due to its age relative to other the platforms on the list and also piggy-backing on the games that came out on their PC application. Because of these reasons it would make sense that SteamOS would have the highest average rating since it can cherry-pick the highest rated games on its market and make it available on its platform. Additionally, we can see that DVD/HD Video games, on average, have the lowest review scores. We believe this is case because these games are usually bundled with a DVD movie. Since the game is just an add-on with the movie the publishers aren’t going to put as much effort to make it a fun game.


### X. Summary - Average Review scores by platform and by publisher
* SteamOS , on average , has the highest review scores
* DVD/ HD Video games, on average, has the lowest review scores


## Global Sales relative to Video Game Rating

### Y. Code Process Walkthrough: Global Sales relative to Video Game Rating
To see if a higher rating from IGN equates to higher global sales, we first made a copy of the review data-set and removed the titles row. To remove anomalies we removed rows that have a release year less than 1996. Additionally, to keep the data-sets uniform we removed consoles that weren’t shared. Afterwards, we then sliced the data to just get the name, platform and global sales sorted by name.
As we were doing this, we found out that the two data-sets we were using didn’t use the same names for each platform, so each platform had to be renamed to keep the data-set consistent. Once, the dataset was uniform we created 3 bar-graphs. One for every rating and every sale, one for every sale with a game rating of 7 or higher and one for a game rating higher than 7 with global sales between 0 – 8 million.


### Z. Global Sales relative to Video Game Rating Code & Graph

```python
#scatter plot of game sales compared to the games review rating
list_of_ratings = reviews.copy() #copy review dataset
list_of_ratings.drop(0,inplace=True) #drop title row

#drop all rows that have a release_year less then 1996
list_of_ratings.drop(list_of_ratings[list_of_ratings.release_year.map(int)<1996].index,inplace=True) 

#drop consoles that are not shared between datasets
list_of_ratings.drop(list_of_ratings[list_of_ratings.platform.isin(['iPad','Macintosh','iPhone','Android',
                                                                    'Linux','Lynx','NeoGeo Pocket Color','Game.Com',
                                                                    'Dreamcast VMU','WonderSwan','Arcade',
                                                                    'Nintendo 64DD','WonderSwan Color',
                                                                    'DVD / HD Video Game','Wireless','Pocket PC',
                                                                    'N-Gage','iPod','Master System','Atari 5200',
                                                                    'TurboGrafx-CD','Sega 32X','Vectrex',
                                                                    'Commodore 64/128','Windows Phone','Web Games',
                                                                    'Windows Surface','Ouya','Atari 2600',
                                                                    'TurboGrafx-16','NES','Genesis','Sega CD'])].index,inplace=True)
#get the subset of data we need and sort it by title
list_of_ratings=list_of_ratings[['title','platform','score']].sort_values(by=['title']).reset_index()
       

list_of_sales = sales.copy() #copy sales dataset
list_of_sales.drop(0,inplace=True) #drop title row

#some data from sales doesn't contain a sales year, we drop it from the dataframe
list_of_sales.drop(list_of_sales[list_of_sales.Year != list_of_sales.Year].index,inplace=True)

#drop all rows that have a year less then 1996
list_of_sales = list_of_sales.drop(list_of_sales[list_of_sales.Year.map(int)<1996].index)

#drop all rows that have a year greater than 2017
list_of_sales = list_of_sales.drop(list_of_sales[list_of_sales.Year.map(int)>2017].index)

#drop consoles that are not shared between datasets
list_of_sales.drop(list_of_sales[list_of_sales.Platform.isin(['PCFX','WS'])].index,inplace=True)

#get the subset of data we need and sort it by title
list_of_sales=list_of_sales[['Name','Platform','Global_Sales']].sort_values(by=['Name']).reset_index()

#to match game titles to pair global sales and score data
#remove all special characters, spaces, make everything lowercase, and set as new column called Merge_Name
list_of_sales['Merge_Name'] = list_of_sales['Name'].copy().str.replace('\W', '')
list_of_ratings['Merge_Name'] = list_of_ratings['title'].copy().str.replace('\W', '')

#Remap all platform names for list_of_ratings
#the sales data sets which list_of_ratings is based on uses the long form platform name instead of shortname
#convert all long form names to short form E.G 'Playstation 2' becomes 'PS2'
list_of_ratings['platform'] = list_of_ratings['platform'].map({"Dreamcast":"DC", "Game Boy":"GB", 
                                                                "Game Boy Advance":"GBA", "Game Boy Color":"GB",
                                                                "GameCube":"GC", "NeoGeo":"NG","New Nintendo 3DS":"3DS",
                                                                "Nintendo 3DS":"3DS", "Nintendo 64":"N64", 
                                                                "Nintendo DS":"DS", "Nintendo DSi":"DS", "PC":"PC", 
                                                                "PlayStation":"PS", "PlayStation 2":"PS2", 
                                                                "PlayStation 3":"PS3","PlayStation 4":"PS4", 
                                                                "PlayStation Portable":"PSP", "PlayStation Vita":"PSV",
                                                                "Saturn":"SAT", "Sega CD":"SCD", "SteamOS":"PC", 
                                                                "Super NES":"SNES", "Wii":"Wii","Wii U":"WiiU", 
                                                                "Xbox":"XB", "Xbox 360":"X360", "Xbox One":"XOne"})

#merge the two dataframes on the Merge_name and platform field
sales_and_ratings = pd.merge(list_of_sales, list_of_ratings, how ="inner", left_on=['Merge_Name','Platform'], right_on = ['Merge_Name','platform'])

#change score & Global_Sales type from string to integer
sales_and_ratings['score']=sales_and_ratings['score'].astype(float)
sales_and_ratings['Global_Sales']=sales_and_ratings['Global_Sales'].astype(float)

test_frame = sales_and_ratings.copy()

#create scatter plot of all data
sales_and_ratings.plot(kind='scatter',x='score',y='Global_Sales', figsize=[20,5], c='Global_Sales', colormap='jet', sharex=False, title="Game Sales Compared to Game Reviews")

#drop rows with less then 7 score and more then 20 million sold
sales_and_ratings = sales_and_ratings.drop(sales_and_ratings[sales_and_ratings.Global_Sales.map(int)>20].index)
sales_and_ratings = sales_and_ratings.drop(sales_and_ratings[sales_and_ratings.score.map(int)<7].index)

#create scatter plot of games with a score better then 7 and with less then 20 million sales
sales_and_ratings.plot(kind='scatter',x='score',y='Global_Sales', figsize=[20,5], c='Global_Sales', colormap='jet', sharex=False, title="Game Sales Compared to Game Reviews")

#drop rows with global sales less then 8 
sales_and_ratings = sales_and_ratings.drop(sales_and_ratings[sales_and_ratings.Global_Sales.map(int)>8].index)

#create scatter plot of games with a score better then 7 and with less then 8 million sales
sales_and_ratings.plot(kind='scatter',x='score',y='Global_Sales', figsize=[20,5], c='Global_Sales', colormap='jet', sharex=False, title="Game Sales Compared to Game Reviews")
```
